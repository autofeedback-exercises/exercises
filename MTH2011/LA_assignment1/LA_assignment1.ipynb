**Example problem**: Solve the system of Equations

$$\begin{cases} 5x+3y-7z = 4 \\ 2x-4y+2z=3\end{cases}$$

We begin by setting up the three symbolic variables we need, $x$, $y$  and $z$. 



```python
import sympy as sy
x,y,z = sy.symbols("x,y,z")
```

Then, we can use the `sy.Eq()` command to build the two equations

```python
Eq1 = sy.Eq(5*x+3*y-7*z, 4)
Eq2 = sy.Eq(2*x-4*y+2*z, 3)
```

Recall that `sy.Eq()` takes two arguments, the left and right hand sides of the equation, respectively. Finally, we use the `sy.solve` command to obtain a solution, and to check it looks correct, we can print it with `sy.pprint`.

```python
q0 = sy.solve([Eq1, Eq2], (x,y))
sy.pprint(q0)
```

Two things to notice in the command to solve

1. The list of equations is enclosed in square brackets.
2. The list of variables for which we want a solution appear in round brackets. Note that in this case we only include two variables ($x$ and $y$), because we only have two, linearly independent equations.
