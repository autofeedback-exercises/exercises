Solve the system of Equations

$$\begin{cases} 5x+3y-7z = 4 \\ 2x-4y+2z=3\end{cases}$$

```python
import sympy as sy

####################################################
# Question 0: worked example

# Set up the symbolic variables
x,y,z = sy.symbols("x,y,z")

# Build the equations
Eq1 = sy.Eq(5*x+3*y-7*z, 4)
Eq2 = sy.Eq(2*x-4*y+2*z, 3)

# Compute the solution and print it. Note the equations are in a list [square brackets] and the requested variables in a so called 'tuple' (round brackets)

q0 = sy.solve([Eq1, Eq2], (x,y))
sy.pprint(q0)
```
